```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

### Строим базисные функции

```python
def calc_basis_function(value, knots, power):
    p = 0
    cur_basis = map(lambda tup: 1. if tup[0] <= value and value < tup[1] else .0, 
                    zip(knots[:-1], knots[1:]))
    #print cur_basis
    cur_basis = np.array(cur_basis)
    while p < power:
        #print cur_basis
        p += 1
        n_basis = len(knots) - p - 1
        next_basis = np.zeros(n_basis)
        for i in range(n_basis):
            if (knots[i+p] - knots[i]) != 0:
                next_basis[i] = (value - knots[i]) / (knots[i+p] - knots[i]) * cur_basis[i]
            else:
                next_basis[i] = 0.0
            if (knots[i+p+1] - knots[i+1]) != 0:
                next_basis[i] += (knots[i+p+1] - value) / (knots[i+p+1] - knots[i+1]) * cur_basis[i+1]
        cur_basis = next_basis
    return cur_basis
```

### Построим пример

```python
knots = sorted((np.random.random(30) * 10).round())
power = 10
support = np.linspace(min(knots), max(knots), num=1000)
results = np.array(map(lambda v: calc_basis_function(v, knots, power), support)).T
for y in results:
    plt.plot(support, y)
plt.show()
plt.plot(support, results.sum(axis=0))
```

### Строим замкнутый Б сплайн

```python
p = 3
control_points = np.random.random([7,2])*10
control_points[:p] = control_points[-p:]
n = len(control_points) - 1
knots = np.linspace(0,10,11)
support = np.linspace(knots[p], knots[-p-1], num=1000)[:-1]
results = np.array(map(lambda v: calc_basis_function(v, knots, p), support)).T
for y in results:
    plt.plot(support, y)
plt.show()
plt.plot(support, results.sum(axis=0))
```

### Используем Алгоритм де Бура чтобы найти контрольные точки для Безье кривых

```python
def deBoor_algo(val, knots, control_points, p):
    h = p
    s = 0
    k = 0
    for i in range(len(knots) - 1):
        if val == knots[i]:
            s += 1
            h -= 1
            k = i
        if knots[i] < val and val < knots[i+1]:
            k = i
            break
    cur_points = control_points[k-p:k-s+1][::]
    gen_points = []
    for r in range(1,h+1):
        next_points = np.zeros([1-s+p-r,2])
        for i in range(k-p+r,k-s+1):
            a_i_r = (val - knots[i]) / (knots[i+p-r+1] - knots[i]) 
            next_points[i-k+p-r] = (1 - a_i_r) * cur_points[i-k+p-r] + a_i_r * cur_points[i+1-k+p-r]
            gen_points.append(next_points[i-k+p-r])
        cur_points = next_points
    return np.array(gen_points)
```

```python
bezier_cp = []
i = 0
for knot in knots[p:-p]:
    gen_points = deBoor_algo(knot, knots, control_points, p)
    bezier_cp.append([gen_points[0], gen_points[2], gen_points[1]])
    i += 1

buf = []
for i in range(len(bezier_cp)):
    buf.append(np.array(bezier_cp[i-1][1:] + bezier_cp[i][:-1]))
bezier_cp = buf
```

### Теперь отобразим тот же сплайн, но не с помощью базисных функций, а через кривые Безье, используя алгоритм Де Кастельжо

```python
def deCasteljau_algo(value, control_points):
    Q = np.copy(control_points)
    for k in range(1,len(control_points)):
        for i in range(len(control_points) - k):
            Q[i] = (1-value)*Q[i] + value*Q[i+1]
    return Q[0]
```

```python
plt.plot(control_points.T[0],control_points.T[1])
for points in bezier_cp[1:]:
    support = np.linspace(0.,1.,100)
    bezier = np.array(map(lambda v: deCasteljau_algo(v, points), support))
    plt.plot(bezier.T[0], bezier.T[1])
    plt.plot(points.T[0], points.T[1])
    plt.scatter(points.T[0], points.T[1], c='r')
```

### сделаем мультик

```python
def plot_spline(control_points, knots, p):
    bezier_cp = []
    i = 0
    for knot in knots[p:-p]:
        gen_points = deBoor_algo(knot, knots, control_points, p)
        bezier_cp.append([gen_points[0], gen_points[2], gen_points[1]])
        i += 1

    buf = []
    for i in range(len(bezier_cp)):
        buf.append(np.array(bezier_cp[i-1][1:] + bezier_cp[i][:-1]))
    bezier_cp = buf
    plt.xlim(2,10)
    plt.ylim(2,10)
    plt.plot(control_points.T[0],control_points.T[1])
    for points in bezier_cp[1:]:
        support = np.linspace(0.,1.,100)
        bezier = np.array(map(lambda v: deCasteljau_algo(v, points), support))
        plt.plot(bezier.T[0], bezier.T[1])
```

```python
plot_spline(control_points, knots, p)
```

```python
vs = np.random.uniform(-2,2,size=control_points.shape)
vs[:p] = vs[-p:]
```

добавим ограничения

```python
def frame_constraints(cur_points, xlims, ylims):
    xlen = xlims[1] - xlims[0]
    ylen = ylims[1] - ylims[0]
    valid = False
    while not valid:
        valid = True
        for p in cur_points:
            if p[0] > xlims[1]:
                valid &= False                
                p[0] = xlims[1] - (p[0] - xlims[1])
            if p[0] < xlims[0]:
                valid &= False
                p[0] = xlims[0] + (xlims[0] - p[0])
            if p[1] > ylims[1]:
                valid &= False
                p[1] = ylims[1] - (p[1] - ylims[1])
            if p[1] < ylims[0]:
                valid &= False
                p[1] = ylims[0] + (ylims[0] - p[1])
    return cur_points
```

```python
def fig_to_np(fig):
    fig.canvas.draw()
    data = np.fromstring(fig.canvas.tostring_rgb(), dtype=np.uint8, sep='')
    data = data.reshape(fig.canvas.get_width_height()[::-1] + (3,))
    return data
```

```python
def frame(t):
    fig = plt.figure(figsize=(10, 10))
    fig.add_subplot(111)
    cur_points = control_points + vs*t
    cur_points = frame_constraints(cur_points, [2,10], [2,10])
    plot_spline(cur_points, knots, p)
    plt.close(fig)
    return fig_to_np(fig)
    
animation = VideoClip(frame, duration=20)
animation.write_gif("bspline.gif", fps=20)
animation.write_videofile("bspline.mp4", fps=20)
```
